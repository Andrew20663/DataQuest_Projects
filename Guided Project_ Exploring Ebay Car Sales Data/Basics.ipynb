## Step 2: Data Exploration 

The next step is to determine what other cleaning tasks that need to be done. In this step I looked for columns where all or almost all values were the same because it can be removed most of the time. Also, numeric data stored as text can also be cleaned and converted. Let's take a look at the data with the describe method for autos. 

```python
autos.describe(include = 'all')
```

For the autos dataframe, we can see that the columns have a lot of NaN values across the board. One key important note is that the seller, offer_type and num_photos have nearly all the same values.  Also, the price and odometer columns are numeric values stored as text. So, we will remove any non-numeric characters and convert the column to a numeric dtype. Let's start by first removing the unecessary columns in our analysis. Then, I will fix the price column and then the odometer column. 

```python
autos['num_photos'].value_counts()
autos['offer_type'].value_counts()
autos['seller'].value_counts()
autos = autos.drop(['num_photos', 'offer_type', 'seller'], axis = 1)
```

### Part A) Fix Price Column 

```python
autos['price'].head()
autos['price'] = autos['price'].str.replace('$', '').str.replace(',', '').astype(float)
autos['price'].head()
```

### Part B) Fix Odometer Column 

```python
autos['odometer'].head()
autos.rename({'odometer':'odometer_km'}, axis = 1 , inplace = True)
autos.columns
```

### Part Bi) Odometer Column Analysis

```python
autos['odometer_km'].head()
autos['odometer_km'] = autos['odometer_km'].str.replace(',', '').str.replace('km','').astype(float)
autos['odometer_km'].head()
```

```python
autos['odometer_km'].unique().shape
print('\n')
print(autos['odometer_km'].describe())
print('\n')
print(autos['odometer_km'].value_counts().head().sort_index(ascending = False))
```

### Ai) Price Analysis

```python
print(autos['price'].unique().shape)
print('\n')
print(autos['price'].describe())
print('\n')
print(autos['price'].value_counts().head(10))
print('\n')
print(autos['price'].value_counts().sort_index().head(10))
print('\n')
autos = (autos[autos['price'].between(1,351000)])
print('\n')
print(autos['price'].describe())

```

From our data, the site seemed to be crawled daily in March and April. Also, the distribution of listings each day seems to be roughly uniform. Also, from the registration year column, it seems that there are odd outliers like the minimum and maximum which suggest the first car was registered in the year 1000 and the latest car was registered in the year 9999.


```python
print(autos['registration_year'].describe())
print('\n')
print(autos['registration_year'].value_counts())
IQR = 2008 - 1999
Outlier_Q1 = 1999 - 1.5 * IQR 
Outlier_Q3 = 2008 + 1.5 * IQR
print(Outlier_Q1, Outlier_Q3)
```

The lowest value I decided to use for comparison was 1986 and 2018 for the highest value because I calculated what the two outlier boundaries were. Anything lower than 1985 is an outlier and anything over 2021 is an outlier as well. Therefore, I will use the interval of comparison from (1986, 2018).

```python
(~autos['registration_year'].between(1986,2018)).sum() / autos.shape[0]
autos = autos[autos['registration_year'].between(1986,2018)]
print(autos['registration_year'].value_counts(normalize = True))
```

```python
top_brands = autos['brand'].value_counts(normalize = True)
top_5_brands = top_brands[top_brands > .05].index
print(top_5_brands)
final_dictionary = {}
for b in top_5_brands: 
    brands = autos[autos['brand'] == b]
    mean_price = brands['price'].mean()
    final_dictionary[b] = float(mean_price)
final_dictionary
```

From here, we can see that Audi, BMW and Mercedes have a higher average price than the other car brands listed. Afterwards, there is a distinct price gap between the other car brands. Opel and Ford are the cheapest while Volkswagen is in the middle. 

```python
mean_mileage = {}
for b in top_5_brands: 
    brands = autos[autos['brand'] == b]
    mean_mile = brands['odometer_km'].mean()
    mean_mileage[b] = float(mean_mile)
mean_mileage
price = pd.Series(final_dictionary).sort_values(ascending = False)
mile = pd.Series(mean_mileage).sort_values(ascending = False)
brand_information = pd.DataFrame(mile, columns = ['Average_Km'])
brand_information['mean_price'] = price
print(brand_information)
```

Overall, we can see that BMW and Mercedes have the highest average km but are not necessarily the highest in average price.
