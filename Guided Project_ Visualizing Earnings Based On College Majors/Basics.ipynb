# Guided Project: Visualizing Earnings Based on College Majors

For this guided project, the goal was to explore how using pandas for plotting was very functional with the dataset on job outcomes of students who graduated college between 2010 and 2012. The original data can be found on the American Community Survey and has aggregate data. The data contains different majors in college, gender diversity, employment rates, median salaries and etc. for its rows. Two questions I wanted to explore were: 

1) Do students in more popular majors make more money?
2) How many majors are predominantly male and female? 
3) Which category of majors have the most students. 

Using pandas, we will answer these three questions where the first question will be answered with a scatterplot, the second will be answered with histograms, and the third will be answered using barplots. In the process as well, we'll remove rows and null values. 

### Step 1: Import Libraries and Clean Data

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
recent_grads = pd.read_csv('recent-grads.csv')
recent_grads.iloc[0]
recent_grads.head()
recent_grads.tail()
recent_grads.describe()
raw_data_count = recent_grads.shape[0]
raw_data_count
recent_grads = recent_grads.dropna()
cleaned_data_count = recent_grads.shape[0]
print('There are', raw_data_count, 'rows')
print('There are', cleaned_data_count, 'rows')
```

### Step 2: Answer Questions

#### Question 1: 

```python
ax = recent_grads.plot(x = 'Sample_size', y = 'Median', kind = 'scatter')
ax.set_xlabel('Sample Size')
ax.set_ylabel('Median')
ax.set_title('Sample Size vs. Median')
plt.show()
ax1 = recent_grads.plot(x = 'Sample_size', y = 'Unemployment_rate', kind = 'scatter')
ax1.set_xlabel('Sample Size')
ax1.set_ylabel('Unemployment Rate')
ax1.set_title('Sample Size vs. Unemployment Rate')
plt.show()
ax2 = recent_grads.plot(x = 'Full_time', y = 'Median', kind = 'scatter')
ax2.set_xlabel('Full Time')
ax2.set_ylabel('Median')
ax2.set_title('Full Time vs. Median')
plt.show()
ax3 = recent_grads.plot(x = 'ShareWomen', y = 'Unemployment_rate', kind = 'scatter')
ax3.set_xlabel('Share Women')
ax3.set_ylabel('Unemployment Rate')
ax3.set_title('Share Women vs. Unemployment Rate')
plt.show()
ax4 = recent_grads.plot(x = 'Men', y = 'Median', kind = 'scatter')
ax4.set_xlabel('Men')
ax4.set_ylabel('Median')
ax4.set_title('Men vs. Median')
plt.show()
ax5 = recent_grads.plot(x = 'Women', y = 'Median', kind = 'scatter')
ax5.set_xlabel('Women')
ax5.set_ylabel('Median')
ax5.set_title('Women vs. Median')
plt.show()
```

From the plots, it is not apparent to say that students in more popular majors make more money. From the Women vs. Median scatterplot, it shows that there is a slight positive correltation to none that subjects that were majority women made more money. Also, there is no link between the number of full-time employees and median salary. 

#### Question 2 

```python
ax1 = recent_grads['Sample_size'].hist(bins = 25, range=(0,500))
ax1.set_title('Sample Size Histogram of Recent Grads')
plt.show()
ax2 = recent_grads['Median'].hist(bins = 25, range=(0,500))
ax2.set_title('Median Histogram of Recent Grads')
plt.show()
ax3 = recent_grads['Employed'].hist(bins = 25, range=(0,500))
ax3.set_title('Employed Histogram of Recent Grads')
plt.show()
ax4 = recent_grads['Full_time'].hist(bins = 25, range=(0,500))
ax4.set_title('Full Time Histogram of Recent Grads')
plt.show()
ax5 = recent_grads['ShareWomen'].hist(bins = 25, range=(0,500))
ax5.set_title('Share Women of Recent Grads')
plt.show()
ax6 = recent_grads['Unemployment_rate'].hist(bins = 25, range=(0,500))
ax6.set_title('Unemployment Rate of Recent Grads')
plt.show()
ax7 = recent_grads['Men'].hist(bins = 25, range=(0,500))
ax7.set_title('Men of Recent Grads')
plt.show()
ax8 = recent_grads['Women'].hist(bins = 25, range=(0,500))
ax8.set_title('Women of Recent Grads')
plt.show()
```
