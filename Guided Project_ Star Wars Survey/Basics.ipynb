# Guided Project: Star Wars Survey

```python
import pandas as pd 
import numpy as np 
star_wars = pd.read_csv('star_wars.csv', encoding = 'ISO-8859-1')
star_wars.head(10)
star_wars.columns
```

```python
star_wars = star_wars[pd.notnull(star_wars['RespondentID'])]
print(star_wars['RespondentID'].isnull().sum())
star_wars.head()
```

```python
yes_no = {'Yes':True, 'No':False}
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)
star_wars.head()
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)
star_wars.head()
```

```python
col_names = {'Which of the following Star Wars films have you seen? Please select all that apply.' : 'seen_1',
            'Unnamed: 4' : 'seen_2', 'Unnamed: 5' : 'seen_3', 'Unnamed: 6' : 'seen_4', 'Unnamed: 7' : 'seen_5',
            'Unnamed: 8' : 'seen_6'}
star_wars = star_wars.rename(col_names, axis = 1)
star_wars.head()
```

```python
diction_1 = {'Star Wars: Episode I  The Phantom Menace': True, np.NaN : False}
diction_2 = {'Star Wars: Episode II  Attack of the Clones': True, np.NaN: False}
diction_3 = {'Star Wars: Episode III  Revenge of the Sith': True, np.NaN: False}
diction_4 = {'Star Wars: Episode IV  A New Hope': True, np.NaN: False}
diction_5 = {'Star Wars: Episode V The Empire Strikes Back':True, np.NaN: False}
diction_6 = {'Star Wars: Episode VI Return of the Jedi': True, np.NaN: False}
star_wars['seen_1'] = star_wars['seen_1'].map(diction_1)
star_wars['seen_2'] = star_wars['seen_2'].map(diction_2)
star_wars['seen_3'] = star_wars['seen_3'].map(diction_3)
star_wars['seen_4'] = star_wars['seen_4'].map(diction_4)
star_wars['seen_5'] = star_wars['seen_5'].map(diction_5)
star_wars['seen_6'] = star_wars['seen_6'].map(diction_6)
star_wars.head() 
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
star_wars = star_wars.rename({'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.' : 'ranking_1'}, axis = 1)
star_wars = star_wars.rename({'Unnamed: 10' : 'ranking_2'}, axis = 1)
star_wars = star_wars.rename({'Unnamed: 11' : 'ranking_3'}, axis = 1)
star_wars = star_wars.rename({'Unnamed: 12': 'ranking_4'}, axis = 1)
star_wars = star_wars.rename({'Unnamed: 13' : 'ranking_5'}, axis = 1)
star_wars = star_wars.rename({'Unnamed: 14' : 'ranking_6'}, axis = 1)
star_wars[star_wars.columns[9:15]]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
star_wars_mean = star_wars[star_wars.columns[9:15]].mean()
print(star_wars_mean)
plt.bar(range(6), star_wars_mean)
plt.xlabel('Ranking Column')
plt.ylabel('mean')
plt.title('Average Rankings for each Ranking Column')
plt.show()
```

```python
star_wars_sum = star_wars[star_wars.columns[3:9]].sum()
print(star_wars_sum)
plt.bar(range(6), star_wars_sum)
plt.title('Seen Sum')
plt.xlabel('Seen Columns')
plt.ylabel('Sum')
plt.show()
```

```python
star_wars_male = star_wars[star_wars['Gender'] == 'Male']
star_wars_female = star_wars[star_wars['Gender'] == 'Female']
print(star_wars_male.head())
print('\n')
print(star_wars_female.head())
```
